<h2 style="color:#c94c4c">Dataset Overview </h2>

The Ford GoBike dataset for February 2019 provides detailed information on the bike-sharing system's usage . The dataset contains over 180,000 bike trips.

The dataset includes information on trip duration, start and end timesof the trips , user type (subscriber or customer), and user demographics such as gender and birth year. The data is provided in CSV format and includes station data as well as users data.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_bike_clean=pd.read_csv('201902_fordgobike_clean.csv')
```

<h2 style="color:#034f84">1. What is the general characteristics of the bike-Sharing system usage?</h2>

Distribution of the trip in February 2019

```python
plt.figure(figsize=[6,4])
default_color = sns.color_palette()[0]
df_bike_clean.groupby('user_type')\
.size().plot(kind='pie',startangle=90,autopct='%1.0f%%',textprops= {'fontsize':10},colors=['lightsteelblue','steelblue'])
plt.ylabel('',size=5)
plt.xlabel('Type of Users')
plt.show()
```

91% of the bike trips were made by registered subscribers of the Ford GoBike program.

```python
default_color=sns.color_palette()[0]
figsize=[8,5]
sns.countplot(data=df_bike_clean,x='member_gender',color=default_color,order=df_bike_clean.member_gender.value_counts().index)
plt.show()
```

The number of male users is more than double the number of female users and others

<h2 style="color:#034f84">2. How long are the trip on average ? And do they wary by user ,type or gender ?</h2>

> You should have at least three visualizations in your presentation,
but feel free to add more if you'd like!

```python
log_binsize = 0.05
bins = 10 ** np.arange(0.05, np.log10(df_bike_clean['duration_min'].max()) + log_binsize,log_binsize)
plt.figure(figsize=[8,5])
#plt.hist(data = df_bike_clean,x ='duration_min',bins = bins)
sns.histplot(data=df_bike_clean,x ='duration_min',bins=bins,color=default_color)


plt.xscale('log')
plt.xticks([5,10,15,30 ,50, 100,300,500, 1e3, 2e3], [5,10,15,30,50,100,300,500, '1k', '2k'])
plt.xlabel('trip duration')
plt.axvline(x=df_bike_clean.duration_min.mean(),color='red')

plt.show()
```
